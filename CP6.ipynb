```
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots()

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(corr, cmap = cmap,annot = True )
```

```
plot_correlation_map(data[["Age", "Survived", "Pclass"]]) # Correlation matrix 
```

```
data["Age"].fillna(data['Age'].mean())
```

```
data.dropna(inplace=True)
```

```
data.head()
```

```
sex = pd.get_dummies(data["Sex"],drop_first=True)
```

```
embarked = pd.get_dummies(data["Embarked"],drop_first=True)
```

```
pclass = pd.get_dummies(data["Pclass"],drop_first=True)
```

```
data = pd.concat([data,pclass,sex,embarked],axis=1)
data.drop(["PassengerId","Pclass","Name","Sex","Ticket","Embarked"],axis=1,inplace=True)
```

```
data.head()
```

```
label = data['Survived']
feat = data.drop(["Survived"], axis=1)
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(feat, label, test_size = 0.2, random_state = 30)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 
dtree = DecisionTreeClassifier(criterion = 'gini', random_state = 0)
dtree.fit(X_train, y_train)
print('Decision Tree Classifier Training Accuracy:', dtree.score(X_test, y_test))

dtree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtree.fit(X_train, y_train)
print('Decision Tree Classifier Training Accuracy:', dtree.score(X_test, y_test))
```

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```
dtree = DecisionTreeClassifier(criterion = 'entropy', max_leaf_nodes= 20, max_depth = 10)
dtree.fit(X_train, y_train)
print('Decision Tree Classifier Training Accuracy:', dtree.score(X_test, y_test))
```

```
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
forest.fit(X_train, y_train)
```

```
print('Random Forest Classifier Training Accuracy:', forest.score(X_test, y_test))
```

```
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100, criterion = 'gini', random_state = 30)
forest.fit(X_train, y_train)
print('Random Forest Classifier Training Accuracy:', forest.score(X_test, y_test))
```
